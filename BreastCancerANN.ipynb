```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/content/data-1.csv')
```

```python
data
```

```python
data.columns
```

```python
data.describe()
```

```python
data.drop('Unnamed: 32',axis=1,inplace=True)
```

```python
data.drop('id',axis=1,inplace=True)
```

```python
data
```

```python
data.isna().sum()
```

```python
data['diagnosis'].replace({'M':1,'B':0},inplace=True)
data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_cols = scaler.fit_transform(data)
scaled_data = pd.DataFrame(scaled_cols,columns=data.columns)
scaled_data
```

```python
X = scaled_data.drop('diagnosis',axis=1)
Y = scaled_data['diagnosis']
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(31,input_shape=(30,),activation='relu'),
    tf.keras.layers.Dense(25,activation='relu'),
    tf.keras.layers.Dense(2,activation='sigmoid'),
])

model.summary()

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(X_train,Y_train,epochs=20)
```

```python
Y_pred = model.predict(X_test)
altered_Y = []
for i in Y_pred:
  if i[1] > 0.5 :
    altered_Y.append(1)
  else:
    altered_Y.append(0)
altered_Y = np.array(altered_Y)
altered_Y
```

```python
from sklearn.metrics import accuracy_score , precision_score , recall_score , f1_score
print(accuracy_score(Y_test,altered_Y))
print(precision_score(Y_test,altered_Y))
print(recall_score(Y_test,altered_Y))
```
